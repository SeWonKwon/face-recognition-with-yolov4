### 방법 2. Google's Open Images Dataset

1.  [데이터 셋 탐색](https://storage.googleapis.com/openimages/web/index.html)




2. [데이터 셋 다운로드](https://github.com/theAIGuysCode/OIDv4_ToolKit) (Code -> Download Zip)

* 설치 : 해당 디렉토리에서 압축을 해제 한다. 또는 git clone

```
git clone https://github.com/theAIGuysCode/OIDv4_ToolKit.git

pip install -r requirements.txt

python main.py downloader --classes Dog Cat --type_csv train --limit 1500

python main.py downloader --classes Dog Cat --type_csv validation --limit 150
``` 


3. 레이블링하기

    1) classes.txt 수정하기  
    2) convert_annotations.py 실행하기  
    3) 하나의 디렉토리로 옮기기
 

```
python convert_annotations.py
```

## 단계 2. 데이터 압축하기 

1. 훈련용 데이터는 train 디렉토리 안에 train.zip 파일로 압축한다.

2. 검증용 데이터는 valid 디렉토리 안에 valid.zip 파일로 압축한다.

2. 구글 드라이브에 업로드 한다.
